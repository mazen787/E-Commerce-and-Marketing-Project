# **Cleaning Time**

```python
print('Checking for extra things in columns name...\n')

datasets = {
    'Customers': customers,
    'Campaigns': campaigns,
    'Ad Spend': ad_spend,
    'Sessions': sessions,
    'Orders': orders,
    'Support': support
}


for name , table in datasets.items():
  print(f'{name} columns')
  for col in table.columns:
    if col != col.strip():
      print(f"Column '{col}' has extra spaces.")
```

```python
customers.columns = customers.columns.str.strip()
campaigns.columns = campaigns.columns.str.strip()
ad_spend.columns = ad_spend.columns.str.strip()
sessions.columns = sessions.columns.str.strip()
orders.columns = orders.columns.str.strip()
support.columns = support.columns.str.strip()
```

```python
def automated_audit(df_name, df):
    print(f"\n{'='*40}")
    print(f" AUDIT REPORT FOR: {df_name}")
    print(f"{'='*40}")

    null_percent = df.isnull().mean() * 100
    high_nulls = null_percent[null_percent > 0]
    if not high_nulls.empty:
        print(f"\n MISSING VALUES FOUND:")
        print(high_nulls.sort_values(ascending=False).to_string(float_format="%.2f%%"))
    else:
        print("\n No missing values found.")


for name, data in datasets.items():
    automated_audit(name, data)
```

```python
print(f"Number of duplicate rows in customers: {customers.duplicated().sum()}")
print(f"Number of duplicate rows in campaigns: {campaigns.duplicated().sum()}")
print(f"Number of duplicate rows in ad_spend: {ad_spend.duplicated().sum()}")
print(f"Number of duplicate rows in sessions: {sessions.duplicated().sum()}")
print(f"Number of duplicate rows in orders: {orders.duplicated().sum()}")
print(f"Number of duplicate rows in support: {support.duplicated().sum()}")
```

```python
# 1. Sessions Table Cleanup
# ==========================================================
# SUMMARY:
# - DROPPED: 'notes' and 'internal_flag' columns (irrelevant/garbage data).
# - FILLED: Missing UTM parameters replaced with 'unknown' for analysis.
# - KEPT: Missing 'customer_id' left as NaN (represents anonymous traffic).
print("Fixing Sessions Table...")

sessions.drop(columns=['notes', 'internal_flag'], inplace=True, errors='ignore')

utm_cols = ['utm_source', 'utm_medium', 'utm_campaign']
for col in utm_cols:
    sessions[col] = sessions[col].fillna('unknown')
```

```python
# 2. Campaigns Table Cleanup
# ==========================================================
# SUMMARY:
# - ACTION: No rows or columns dropped.
# - KEPT: Missing 'end_date' left as NaN (indicates ongoing/active campaigns).
# No code needed here; decision is to leave NaN as is.
```

```python
# 3. Ad Spend Table Cleanup
# ==========================================================
# SUMMARY:
# - DROPPED: Rows where 'campaign_id' is missing (unattributable spend).
# - REASON: Small percentage (<0.5%), removing to maintain data model integrity.

ad_spend.dropna(subset=['campaign_id'], inplace=True)
```

```python
# 4. Customers Table Cleanup
# ==========================================================
# SUMMARY:
# - FILLED: Missing 'gender' replaced with 'Unknown'.
# - KEPT: Missing 'birth_date' and 'email' left as NaN (missing info but valid customer records).

customers['gender'] = customers['gender'].fillna('Unknown')
```

```python
# 5. Orders Table Cleanup
# ==========================================================
# SUMMARY:
# - FILLED: Missing 'session_id' replaced with 'offline_or_missing'.
# - KEPT: Missing 'customer_id' left as NaN (Guest Checkouts).
# - NOTE: No rows dropped to preserve total revenue figures.

orders['session_id'] = orders['session_id'].fillna('offline_or_missing')
```

```python
# 6. Support Table Cleanup
# ==========================================================
# SUMMARY:
# - ACTION: No changes applied.
# - KEPT: Missing IDs and CSAT scores left as NaN (General inquiries or user declined to rate).
# Decision: Keep NaNs statistically accurate.
```

```python
print("FULL CATEGORICAL SCAN (Checking columns with < 50 unique values)...\n")

for table_name, df in datasets.items():
    print(f" TABLE: {table_name}")
    print("=" * 40)

    text_cols = df.select_dtypes(include=['object']).columns

    found_categorical = False
    for col in text_cols:
        unique_count = df[col].nunique()

        if unique_count < 50:
            found_categorical = True
            print(f"Column: '{col}' ({unique_count} unique values)")
            print(f"   Values: {df[col].unique()}")
            print("-" * 20)

    if not found_categorical:
        print("   (No categorical columns found needing check)")
    print("\n")
```

```python
print("Starting Advanced Standardization Phase...")

# ==============================================================================
# 1. Case Standardization and Whitespace Removal (Initial Cleanup)
# ==============================================================================
# This is the most important step; it converts MOBILE to mobile and FB to fb
str_cols = ['traffic_source', 'utm_source', 'device_type', 'payment_method']
for col in str_cols:
    # We check first to ensure the column exists to avoid errors
    if col in sessions.columns:
        sessions[col] = sessions[col].str.lower().str.strip()
    if col in orders.columns:
        orders[col] = orders[col].str.lower().str.strip()

print("Step 1: Lowercase applied.")
```

```python
# 2. Traffic Source Standardization Dictionary (The Traffic Map)
# ==============================================================================
# Here we tell Python: "Any of these variations should be renamed to this"
traffic_mapping = {
    # Google Variations
    'gads': 'google',
    'google ads': 'google',
    'google search': 'google',
    'googleads': 'google',
    'organic search': 'google',
    'goog': 'google',

    # Facebook Variations
    'fb': 'facebook',
    'meta': 'facebook',
    'facebook ads': 'facebook',

    # Instagram Variations
    'ig': 'instagram',
    'instagram ads': 'instagram',

    # TikTok Variations
    'tik tok': 'tiktok',
    'tiktok ads': 'tiktok',

    # Direct/None Variations
    'direct': 'direct',
    'none': 'direct',
    'unknown': 'unknown' # What we filled previously
}

# Apply standardization to traffic_source and utm_source
sessions['traffic_source'] = sessions['traffic_source'].replace(traffic_mapping)
sessions['utm_source'] = sessions['utm_source'].replace(traffic_mapping)

print("Step 2: Traffic Sources Standardized (Google, Facebook, etc.)")
```

```python
# 3. Payment Method Standardization Dictionary (The Payment Map)
# ==============================================================================
payment_mapping = {
    'cod': 'cash on delivery',
    'cash': 'cash on delivery',
    'card': 'credit card',
    'visa': 'credit card',
    'mastercard': 'credit card',
    'wallet': 'digital wallet',
    'instapay': 'instapay'
}

# Apply the mapping to standardize payment methods in the orders dataframe
orders['payment_method'] = orders['payment_method'].replace(payment_mapping)

print("Step 3: Payment Methods Standardized.")
```

```python
# 4.Final Verification
# ==============================================================================
print("\n CLEANED DATA SNAPSHOT:")
print("--------------------------------")
print("Traffic Sources now looks like:")
print(sessions['traffic_source'].value_counts())
print("\nPayment Methods now looks like:")
print(orders['payment_method'].value_counts())
print("\nDevice Types now looks like:")
print(sessions['device_type'].value_counts())
```

```python
print(" AUDIT: Checking Email Quality...")

# 1. Detect Invalid Formats
# Condition: Email must contain '@' and a '.'
invalid_emails = customers[~customers['email'].astype(str).str.contains('@') | ~customers['email'].astype(str).str.contains(r'\.')]
print(f"Found {len(invalid_emails)} invalid email formats (missing @ or .).")

# Display a sample for verification
if len(invalid_emails) > 0:
    print(invalid_emails['email'].head())

# 2. Detect Duplicates
# Identifying emails shared across different customer_ids
duplicate_emails = customers[customers.duplicated(subset=['email'], keep=False)]
print(f"Found {len(duplicate_emails)} customers sharing the same email address.")
```

```python
print("FIXING: Cleaning Emails...")

# Intelligent Check function:
# If value is empty (NaN) -> return None
# If value is text but missing '@' or '.' -> return None (Delete/Invalidate)
# Otherwise -> return the email as is
def clean_email(email):
    if pd.isna(email): # If it's already empty
        return None

    email_str = str(email).strip() # Remove whitespaces

    # Ensure essential email markers exist
    if '@' in email_str and '.' in email_str:
        return email_str
    else:
        return None # Invalid format

# Apply the function to the column
customers['email'] = customers['email'].apply(clean_email)

# Verify the count of nulls
invalid_now = customers['email'].isnull().sum()
print(f"Email column cleaned. Total valid emails retained. (Null count: {invalid_now})")
```

```python
print("AUDIT: Checking Session Durations...")

# 1. Calculate duration in minutes
# Formula: (End Timestamp - Start Timestamp) / 60 seconds
sessions['duration_minutes'] = (sessions['session_end_ts'] - sessions['session_start_ts']).dt.total_seconds() / 60

# 2. Detect Anomalies
# Case 1: Negative duration (Time traveling!)
negative_time = sessions[sessions['duration_minutes'] < 0]

# Case 2: Sessions longer than 6 hours (360 minutes)
long_sessions = sessions[sessions['duration_minutes'] > 360]

print(f"Found {len(negative_time)} sessions with negative duration (Impossible!).")
print(f"Found {len(long_sessions)} sessions longer than 6 hours (Likely Bots/Errors).")
```

```python
print("FIXING: Removing Session Outliers...")

# We retain only logical sessions (Duration >= 0 and <= 6 hours)
original_count = len(sessions)

sessions = sessions[
    (sessions['duration_minutes'] >= 0) &
    (sessions['duration_minutes'] <= 360)
]

dropped_count = original_count - len(sessions)

print(f"Removed {dropped_count} bad sessions. Data is now logical.")
```

```python
print("\n AUDIT: Checking Math Accuracy in Orders...")

# 1. Convert columns to numeric (to avoid data type errors)
cols = ['subtotal', 'shipping_fee', 'tax_amount', 'discount_amount', 'total_amount']
for c in cols:
    orders[c] = pd.to_numeric(orders[c], errors='coerce').fillna(0)

# 2. Calculate the correct sum (The accounting formula)
orders['correct_total'] = orders['subtotal'] + orders['shipping_fee'] + orders['tax_amount'] - orders['discount_amount']

# 3. Comparison (Rounding to 2 decimal places to avoid floating-point precision issues)
# We check if the recorded Total Amount differs from the calculated Correct Total
orders['math_error'] = round(orders['total_amount'], 2) != round(orders['correct_total'], 2)

error_count = orders['math_error'].sum()
print(f"Found {error_count} orders where Total Amount is WRONG.")

# Show a sample of orders with errors for verification
if error_count > 0:
    print("\nSample of wrong calculations:")
    print(orders[orders['math_error']][['order_id', 'total_amount', 'correct_total']].head())
```

```python
print(" RE-AUDITING: Separating 'Rounding Noise' from 'Real Errors'...")

# Calculate the Absolute Difference
orders['diff'] = abs(orders['total_amount'] - orders['correct_total'])

# 1. Rounding Noise (Differences less than 1.00)
rounding_errors = orders[(orders['diff'] > 0) & (orders['diff'] < 1)]
print(f" Found {len(rounding_errors)} orders with tiny rounding diffs (Ignore these as 'Noise').")

# 2. Significant Errors (Differences of 1.00 or more)
real_errors = orders[orders['diff'] >= 1]
print(f"Found {len(real_errors)} orders with MAJOR calculation errors (The real problem!).")

if len(real_errors) > 0:
    print("\nSample of REAL errors:")
    print(real_errors[['order_id', 'total_amount', 'correct_total', 'diff']].head())
```

```python
print("FIXING: Overwriting Total Amount with Correct Values...")

# Overwriting the original column with the correctly calculated values for the entire dataset
orders['total_amount'] = orders['correct_total']

# Cleanup: Removing the auxiliary columns used during the audit
orders.drop(columns=['correct_total', 'math_error', 'diff'], inplace=True, errors='ignore')

print("Fixed ALL calculation errors. Financial data is now 100% accurate.")
```

```python
print("\nFINAL FIX: Handling Orphan Orders...")

# 1. Get all valid customer IDs from the customers table
valid_ids = customers['customer_id'].unique()

# 2. Identify orders linked to non-existent customer IDs
orphan_mask = (~orders['customer_id'].isin(valid_ids)) & (orders['customer_id'].notnull())
orphan_count = orphan_mask.sum()

print(f" Found {orphan_count} orders linked to non-existent customers.")

# 3. Clean up orphans (Set invalid IDs to None/Null)
if orphan_count > 0:
    orders.loc[orphan_mask, 'customer_id'] = None
    print("Set invalid customer_ids to Null to save Revenue.")
else:
    print("No orphans found.")
```

```python
# 2. Detecting "Negative Financials"
# Products and shipping costs cannot be negative values
print("\n Checking Negative Money:")
financial_cols = ['subtotal', 'shipping_fee', 'tax_amount', 'total_amount']
neg_orders = orders[(orders[financial_cols] < 0).any(axis=1)]

if len(neg_orders) > 0:
    print(f"    ALERT: Found {len(neg_orders)} orders with negative values!")
    # Fix: Convert to positive (assuming they are sign errors)
    for col in financial_cols:
        orders[col] = orders[col].abs()
    print("      Fixed: Converted negative values to absolute (positive).")
else:
    print("   Orders: No negative financial values.")
```

```python
# 3. Campaign Logic Audit (End Date < Start Date)
# Campaigns that ended before they actually started
invalid_campaigns = campaigns[campaigns['end_date'] < campaigns['start_date']]

if len(invalid_campaigns) > 0:
    print(f"    ALERT: Found {len(invalid_campaigns)} campaigns ending before they start!")
    # Solution: Swap dates or delete them
    # Here we drop them because this data is logically inconsistent
    campaigns = campaigns.drop(invalid_campaigns.index)
    print("       Fixed: Removed invalid campaigns.")
else:
    print("   Campaigns: All dates look logical.")
```

```python
# 4. CSAT Scores Audit (Range: 1 to 5)
print("\n4 Checking CSAT Logic (1-5):")
# Are there any ratings greater than 5 or less than 1?
invalid_scores = support[
    (support['csat_score'] > 5) |
    (support['csat_score'] < 1)
]
if len(invalid_scores) > 0:
    print(f"    ALERT: Found {len(invalid_scores)} invalid CSAT scores!")
    # Fix: Set them to NaN (Not a Number)
    support.loc[invalid_scores.index, 'csat_score'] = None
    print("       Fixed: Set invalid scores to Null.")
else:
    print("    Support: CSAT scores are valid (1-5 or Null).")

print("\n FINAL STATUS: DATA IS 100% CLEAN.")
```
